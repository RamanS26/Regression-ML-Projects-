```python
from sklearn.datasets import make_regression # Helps us to create a sample dataset for regression
import numpy as np
import pandas as pd

import plotly.express as px # mathplotlib creates static plots i.e images whereas plotly creates interactive plots
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
x,y = make_regression(n_samples = 100, n_features = 2,n_targets=1)
```

```python
x
```

```python
y
```

```python
df = pd.DataFrame({'feature1':x[:,0],'feature2':x[:,1],'target':y})
```

```python
df.shape
```

```python
df.head()
```

```python
fig = px.scatter_3d(df,x='feature1',y='feature2',z='target')
```

```python
fig.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2) #0.2 = 20% and remaining 80% is for training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print('Mean_absolute_error',mean_absolute_error(y_test,y_pred))
print('Mean_Squared_error',mean_squared_error(y_test,y_pred))
print("R2_score",r2_score(y_test,y_pred))
```

```python
X = np.linspace(-5,5,10)
Y = np.linspace(-5,5,10)
XGrid,YGrid = np.meshgrid(Y,X)

final = np.vstack((XGrid.ravel().reshape(1,100),YGrid.ravel().reshape(1,100))).T

z_final =lr.predict(final).reshape(10,10)

z=z_final
```

```python
fig = px.scatter_3d(df,x = 'feature1',y='feature2',z='target')
fig.add_trace(go.Surface(x=X,y=Y,z=z))
fig.show()
```
